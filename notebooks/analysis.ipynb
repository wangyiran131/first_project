The goal of this project is to use the College Scorecard data provided by the US government to visualize the trend in undergraduate population of UC Davis.  This task is simpler than most Data Science projects, but it is enough to demonstrate the DS workflow.

We will focus on the Undergraduate population at UC Davis, which is the UGDS variable in the College Scorecard.  Each file contains a different school year, and we will represent the school year with the year in the Fall (so that 2000-2001 is represented as 2000).

```python
import matplotlib.pyplot as plt
import pickle
```

```python
## Import summary data
datafile = '../data/Davis_UG.dat'
UG, years = pickle.load(open(datafile,'rb')) 
```

```python
## Plot the summary data
plt.plot(UG,years)
plt.xlabel('year')
_ = plt.ylabel('UG pop.')
```

**Figure 1.** The undergraduate population of UC Davis by year from 1995-2017.  Notice the slight drop in enrollment in 2003 and then steady and growing increase since 2005.
